```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
for fname in os.listdir('../data'):
    print('File: {}'.format(fname))
    x = np.load(os.path.join('../data', fname))
    
    has_nans = np.isnan(x).any()
    print("Array contains NaNs:", has_nans)
    if has_nans:
        print(f'{np.isnan(x).sum()} nans')
    print(f'mean : {x.mean()}')
    print(f'min : {x.min()}')
    print(f'max : {x.max()}')
    plt.imshow(x)
    plt.show()
    print('======================================================')
```
